# Convert everything to pandas, merge, save.

It is convenient to merge the datasets into a single flat file of features and rows.

```python
df_a = a_tab2.to_pandas()
df_b = b_tab2.to_pandas()
df_c = c_tab3.to_pandas()
```

Clean the datasets:
- Replace nonstandard NaN representations with IEEE standard `NaN`
- Rename columns if necessary
- Map categorical values to a pandas `Categorical`

```python
#df_a.head()
```

```python
for col in ['memb', 'LC']:
    df_a[col] = df_a[col].astype('category')
```

That one looks good.  Let's keep only the columns that were new in paper II.

```python
new_in_paper_II = ['Per', 'dd', 'ddm', 'shch', 'beat', 'cpeak',
                   'resc', 'resd', 'dscu', 'cloud']
```

```python
df_b = df_b[['EPIC'] + new_in_paper_II]
```

```python
df_b.columns
```

```python
#df_b.head()
```

```python
for col in ['Per', 'dd', 'ddm','shch', 'beat', 'cpeak', 'resc', 'resd', 'dscu', 'cloud']:
    df_b[col] = df_b[col].astype('category')
```

Try to merge these two dataframes to see if the periods match up.

```python
df_b.columns
```

```python
df_ab = pd.merge(df_a.rename(columns={'Per':'Per_lit'}),
                 df_b.rename(columns={'Per':'Per_cat'}),
                 how='outer', on='EPIC')
```

*Spot check:* Did the merge produce any rows in only one or the other dataframes? Nope! :)

```python
#df_ab.head()
```

Clean the Stauffer et al. table.

```python
df_c.head()
```

```python
df_c[df_c.isin([-9.99, -9.0])] = np.NaN
```

```python
df_c['Name'][df_c['Name']=='x'] = np.NaN
```

We only want to keep information for the periodic sources from Rebull et al. 2016b.

```python
df_abc = pd.merge(df_ab.rename(columns={'Name':'Name_IAU', 'Ksmag':'Ksmag_R16'}),
                  df_c.rename(columns={'Name':'Name_adopt','Ksmag':'Ksmag_S16'}), how='left', on='EPIC')
```
